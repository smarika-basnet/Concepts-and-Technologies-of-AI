**2. Build a Neural Network Model [15]**

• Design and implement a Neural Network–based
classifier (e.g., Multi-Layer Perceptron).

---



```python
# ================== NEURAL NETWORK MODEL WITH SMOTE ==================

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE

# ---------- A. PREPROCESSING ----------
X = train_df.drop(columns=['y', 'duration'])   # drop leakage feature
y = train_df['y'].map({'no':0, 'yes':1})

categorical_cols = X.select_dtypes(include=['object']).columns
numeric_cols = X.select_dtypes(include=['int64','float64']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# ---------- B. TRAIN/TEST SPLIT ----------
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    stratify=y,
                                                    random_state=42)

# ---------- C. HANDLE IMBALANCE (SMOTE) ----------
# Fit preprocessing on train set first
X_train_pre = preprocessor.fit_transform(X_train)
X_test_pre = preprocessor.transform(X_test)

# Apply SMOTE only on training set
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_pre, y_train)

print("Before SMOTE:", np.bincount(y_train))
print("After SMOTE:", np.bincount(y_train_res))

# ---------- D. MODEL ----------
mlp = MLPClassifier(hidden_layer_sizes=(64,32),
                    activation='relu',
                    solver='adam',
                    max_iter=300,
                    random_state=42)

# ---------- E. TRAINING ----------
mlp.fit(X_train_res, y_train_res)

# ---------- F. EVALUATION ----------
y_train_pred = mlp.predict(X_train_res)
y_test_pred = mlp.predict(X_test_pre)

print("Training Accuracy:", accuracy_score(y_train_res, y_train_pred))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))

print("\nConfusion Matrix (Test):\n", confusion_matrix(y_test, y_test_pred))
print("\nClassification Report (Test):\n", classification_report(y_test, y_test_pred))
```

**Neural Network Architecture:**

The Multi-Layer Perceptron (MLP) classifier is configured as follows:

**Architecture:**
- **Input Layer:** Accepts preprocessed features (scaled numerical + one-hot encoded categorical)
- **Hidden Layer 1:** 64 neurons with ReLU activation function
- **Hidden Layer 2:** 32 neurons with ReLU activation function
- **Output Layer:** 1 neuron with sigmoid activation (binary classification)

**Loss Function:**
- Binary Cross-Entropy (log loss)
- Measures the difference between predicted probabilities and actual labels

**Optimization Algorithm:**
- Adam (Adaptive Moment Estimation)
- Learning rate: 0.001 (default)
- Combines benefits of momentum and RMSProp for efficient training

**Training Configuration:**
- Maximum iterations: 300
- Random state: 42 (for reproducibility)

**Design Rationale:**
The architecture uses decreasing layer sizes (64 → 32) to create a funnel effect, gradually compressing information toward the binary output. ReLU activation was chosen to prevent vanishing gradients and speed up training.

```python
# ================== PRIMARY MODELS ==================

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# ---------- A. Logistic Regression ----------
log_reg = LogisticRegression(max_iter=500, random_state=42)

# Fit on training data
log_reg.fit(X_train_pre, y_train)

# Predictions
y_pred_log_train = log_reg.predict(X_train_pre)
y_pred_log_test = log_reg.predict(X_test_pre)

print("Logistic Regression Training Accuracy:", accuracy_score(y_train, y_pred_log_train))
print("Logistic Regression Test Accuracy:", accuracy_score(y_test, y_pred_log_test))
print("\nClassification Report (Test - Logistic Regression):\n", classification_report(y_test, y_pred_log_test))

# ---------- B. Decision Tree ----------
dtree = DecisionTreeClassifier(max_depth=10, random_state=42)

# Fit on training data
dtree.fit(X_train_pre, y_train)

# Predictions
y_pred_dt_train = dtree.predict(X_train_pre)
y_pred_dt_test = dtree.predict(X_test_pre)

print("Decision Tree Training Accuracy:", accuracy_score(y_train, y_pred_dt_train))
print("Decision Tree Test Accuracy:", accuracy_score(y_test, y_pred_dt_test))
print("\nClassification Report (Test - Decision Tree):\n", classification_report(y_test, y_pred_dt_test))
```

**Conclusion**

Both models achieve similar overall accuracy (~89%).
- **Logistic Regression** is more precise but has very low recall for the minority class.
- **Decision Tree** improves recall for the minority class, which is important in imbalanced datasets, though precision is lower.
- **Decision Tree is considered slightly better overall** because it balances accuracy with improved recall for the minority class, making it more useful in scenarios where detecting positives is critical.
 ---
 *Note:* The dataset is highly imbalanced (≈88% "no", 12% "yes").
Balancing the data using SMOTE was applied for the neural network model, which improved recall for the minority class.
For classical models, initial training was done on the imbalanced dataset to establish a baseline.

Follow the steps below to perform this task:

```python
# ================== HYPERPARAMETER OPTIMIZATION ==================

from sklearn.model_selection import GridSearchCV

# ---------- A. Logistic Regression ----------
param_grid_log = {
    'C': [0.01, 0.1, 1, 10],          # regularization strength
    'solver': ['liblinear', 'lbfgs']  # optimization algorithms
}

grid_log = GridSearchCV(LogisticRegression(max_iter=500, random_state=42),
                        param_grid_log,
                        cv=5,
                        scoring='f1_macro',
                        n_jobs=-1)

grid_log.fit(X_train_pre, y_train)

print("Best Logistic Regression Params:", grid_log.best_params_)
print("Best CV Score (Logistic Regression):", grid_log.best_score_)

# ---------- B. Decision Tree ----------
param_grid_dt = {
    'max_depth': [5, 10, 15, None],   # tree depth
    'min_samples_split': [2, 5, 10],  # minimum samples to split
    'criterion': ['gini', 'entropy']  # splitting criteria
}

grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42),
                       param_grid_dt,
                       cv=5,
                       scoring='f1_macro',
                       n_jobs=-1)

grid_dt.fit(X_train_pre, y_train)

print("Best Decision Tree Params:", grid_dt.best_params_)
print("Best CV Score (Decision Tree):", grid_dt.best_score_)
```

Hyperparameter tuning was performed using **GridSearchCV** with 5-fold cross-validation to identify the best parameter settings for both Logistic Regression and Decision Tree models. The evaluation metric used was **F1-macro**, which balances performance across both majority and minority classes.

**Logistic Regression**

- **Interpretation:** Logistic Regression performed best with moderate regularization (C=1) and the lbfgs solver. The CV score indicates balanced performance, though recall for the minority class remains a challenge.

---

**Decision Tree**
- **Hyperparameters tuned:**
  - criterion (splitting criterion: gini or entropy)
  - max_depth (maximum depth of the tree)
  - min_samples_split (minimum samples required to split a node)
- **Best Parameters:** {'criterion': 'gini', 'max_depth': 15, 'min_samples_split': 2}
- **Best Cross-Validation Score (F1-macro):** 0.640
- **Interpretation:** The Decision Tree achieved its best performance with a depth of 15, using the Gini criterion and allowing splits with at least 2 samples. This configuration balances complexity and generalization, improving minority class detection compared to Logistic Regression.

**Conclusion**
