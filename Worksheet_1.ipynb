**Problem - 4: Matrix Operations:**

---



For the following arrays:

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

Prove following with Numpy:

---



a) Prove A.A^-1 = I

```python
A_inv = np.linalg.inv(A)
print("Inverse of A =\n", A_inv)

I = np.dot(A, A_inv)
print("A · A_inv =\n", I)

I_expected = np.eye(2)
print("Expected Identity matrix =\n", I_expected)

print("Is A·A_inv equal to I? ", np.allclose(I, I_expected))

```

b) Prove AB ̸= BA.

```python
AB = np.dot(A, B)
print("AB =\n", AB)

BA = np.dot(B, A)
print("BA =\n", BA)

print("Is AB equal to BA? ", np.array_equal(AB, BA))
```

c) Prove (AB)^T = B^TA^T

```python
AB = np.dot(A, B)
AB_T = AB.T
print("(AB)^T =\n", AB_T)

B_T = B.T
A_T = A.T

BTA_T = np.dot(B_T, A_T)
print("B^T · A^T =\n", BTA_T)

print("Is (AB)^T equal to B^T·A^T? ", np.array_equal(AB_T, BTA_T))
```

d) Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9

```python
# Coefficient matrix
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Constant matrix
B = np.array([-1, -3, 9])

# Find inverse of A
A_inv = np.linalg.inv(A)

# Solve for X
X = np.dot(A_inv, B)
print("Solution [x, y, z] =", X)

```

**4.2 Experiment: How Fast is Numpy?**

---



a) Element-wise Addition:

• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import time

N = 1000000  # size

# Python lists
list1 = list(range(N))
list2 = list(range(N))

start = time.time()
sum_list = [list1[i] + list2[i] for i in range(N)]
end = time.time()
print("Time with Python lists:", end - start, "seconds")

# NumPy arrays
arr1 = np.arange(N)
arr2 = np.arange(N)

start = time.time()
sum_array = arr1 + arr2
end = time.time()
print("Time with NumPy arrays:", end - start, "seconds")
```

b) Element-wise Multiplication

• Using Python Lists, perform element-wise multiplication of two lists of size 1, 000, 000. Mea-
sure and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import time

N = 1000000  # size of lists/arrays

# Python lists
list1 = list(range(N))
list2 = list(range(N))

start = time.time()
mul_list = [list1[i] * list2[i] for i in range(N)]
end = time.time()
print("Time with Python lists:", end - start, "seconds")

# NumPy arrays
arr1 = np.arange(N)
arr2 = np.arange(N)

start = time.time()
mul_array = arr1 * arr2
end = time.time()
print("Time with NumPy arrays:", end - start, "seconds")
```

c) Dot Product

• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import time

N = 1000000  # size of lists/arrays

# Python lists
list1 = list(range(N))
list2 = list(range(N))

start = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(N))
end = time.time()
print("Time with Python lists:", end - start, "seconds")
print("Dot product (Python lists) =", dot_list)

# NumPy arrays
arr1 = np.arange(N)
arr2 = np.arange(N)

start = time.time()
dot_array = np.dot(arr1, arr2)
end = time.time()
print("Time with NumPy arrays:", end - start, "seconds")
print("Dot product (NumPy arrays) =", dot_array)
```

d) Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure
and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

```python
import time
import random

N = 1000  # size of matrices

# Python lists
matrix1 = [[random.randint(0, 10) for _ in range(N)] for _ in range(N)]
matrix2 = [[random.randint(0, 10) for _ in range(N)] for _ in range(N)]

# Matrix multiplication
start = time.time()
result_list = [[0]*N for _ in range(N)]
for i in range(N):
    for j in range(N):
        for k in range(N):
            result_list[i][j] += matrix1[i][k] * matrix2[k][j]
end = time.time()
print("Time with Python lists:", end - start, "seconds")

# NumPy arrays
arr1 = np.random.randint(0, 11, size=(N, N))
arr2 = np.random.randint(0, 11, size=(N, N))

# Matrix multiplication
start = time.time()
result_array = np.dot(arr1, arr2)
end = time.time()
print("Time with NumPy arrays:", end - start, "seconds")
```
