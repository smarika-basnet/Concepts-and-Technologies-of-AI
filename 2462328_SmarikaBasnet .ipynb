**6. Adding HDI Category Column:**

• Create a new column HDI Category that classifies each country into one of the four official
Human Development Index groups. The classification should be based on the HDI value for the
year 2022. Use the given categories and thresholds defined by the United Nations Development
Programme (UNDP)

After creating the new column:

• verify that all countries are classified correctly,

• ensure the updated dataframe includes the new category column.

• Save the final dataframe as HDI category added.csv and include this file in your final
submission.

---



```python
# Define function to categorize HDI
def categorize_hdi(hdi_value):
    if hdi_value < 0.550:
        return 'Low'
    elif 0.550 <= hdi_value <= 0.699:
        return 'Medium'
    elif 0.700 <= hdi_value <= 0.799:
        return 'High'
    else:  # hdi >= 0.800
        return 'Very High'

# Create new column 'HDI_Category'
hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(categorize_hdi)

# Verify classification: show counts per category
category_counts = hdi_2022_df['HDI_Category'].value_counts()
print("Number of countries in each HDI category:")
print(category_counts)

# Display first 10 rows including the new column
hdi_2022_df[['country', 'hdi', 'HDI_Category']].head(10)

# Save the updated dataframe to CSV
hdi_2022_df.to_csv('/content/HDI_category_added.csv', index=False)
```

-> Each country in the 2022 dataset has been classified into one of four HDI categories: Low, Medium, High, or Very High, based on the official UNDP thresholds.

-> The distribution of countries across categories can be seen in the value counts.

-> The updated dataframe has been saved as 'HDI_category_added.csv'.

```python
# Load the filtered dataset for 2020-2022
hdi_2020_2022_df = pd.read_csv('HDI_problem1B.csv')

# 1. Check for missing values in essential columns
missing_essential = hdi_2020_2022_df[['hdi', 'country', 'year']].isnull().sum()
print("Missing values in essential columns:\n", missing_essential)

# 2. Inspect country names for inconsistencies (sample check)
sample_countries = hdi_2020_2022_df['country'].unique()[:10]
print("\nSample country names:", sample_countries)

# 3. Check for duplicates
duplicates_count = hdi_2020_2022_df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates_count)

# 4. Check numeric columns for non-numeric symbols (if any)
numeric_cols = ['hdi', 'life_expectancy', 'pop_millions', 'gross_inc_percap']
for col in numeric_cols:
    non_numeric = pd.to_numeric(hdi_2020_2022_df[col], errors='coerce').isnull().sum()
    print(f"Non-numeric entries in '{col}':", non_numeric)

# 5. Cleaning steps:

# Drop rows with missing values in essential columns
hdi_2020_2022_df.dropna(subset=['hdi', 'country', 'year'], inplace=True)

# Convert numeric columns to numeric dtype
hdi_2020_2022_df[numeric_cols] = hdi_2020_2022_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Remove duplicate rows
hdi_2020_2022_df.drop_duplicates(inplace=True)

# Reset index
hdi_2020_2022_df.reset_index(drop=True, inplace=True)

# Verify dataset after cleaning
print("\nDataset shape after cleaning:", hdi_2020_2022_df.shape)
print("\nMissing values after cleaning:\n", hdi_2020_2022_df[['hdi', 'country', 'year']].isnull().sum())
```

**2. Composite Development Score:**

• Create a new metric called Composite Score using the formula:

Composite Score = 0.30 × Life Expectancy Index + 0.30 × GNI per Capita Index}
Here: Life Expectancy Index → "life expectancy" and GNI per Capita Index → "gross inc percap"

• Rank South Asian countries based on Composite Score.

• Plot the top 5 countries in a horizontal bar chart.

• Compare the ranking of countries by Composite Score with their HDI ranking and discuss any
differences.

---



```python
# Take the latest year for each country
hdi_latest = hdi_south_asia[hdi_south_asia['year'] == 2022]

# Compute Composite Score
hdi_latest.loc[:, 'Composite_Score'] = 0.30 * hdi_latest['life_expectancy'] + 0.30 * hdi_latest['gross_inc_percap']

# Sort by Composite Score
hdi_latest_sorted = hdi_latest.sort_values('Composite_Score', ascending=False)

# Select top 5 countries
top5_countries = hdi_latest_sorted.head(5)

# Plot horizontal bar chart
plt.figure(figsize=(10,6))
sns.barplot(
    data=top5_countries,
    y='country',
    x='Composite_Score',
    palette='viridis'
)
plt.title('Figure 5: Top 5 South Asian Countries by Composite Score (2022)', fontsize=14)
plt.xlabel('Composite Score', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.xlim(0, top5_countries['Composite_Score'].max()*1.1)
plt.tight_layout()
plt.savefig('Figure5_SA_Countries_Composite_Score.png', dpi=300, bbox_inches='tight')
plt.show()
```

**3. Outlier Detection:**

• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.

• Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

• Discuss why the identified countries stand out as outliers.

---



```python
# HDI outliers
Q1_hdi = hdi_south_asia['hdi'].quantile(0.25)
Q3_hdi = hdi_south_asia['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi
hdi_outliers = hdi_south_asia[(hdi_south_asia['hdi'] < Q1_hdi - 1.5*IQR_hdi) |
                              (hdi_south_asia['hdi'] > Q3_hdi + 1.5*IQR_hdi)]

# GNI per Capita outliers
Q1_gni = hdi_south_asia['gross_inc_percap'].quantile(0.25)
Q3_gni = hdi_south_asia['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni
gni_outliers = hdi_south_asia[(hdi_south_asia['gross_inc_percap'] < Q1_gni - 1.5*IQR_gni) |
                              (hdi_south_asia['gross_inc_percap'] > Q3_gni + 1.5*IQR_gni)]
```

**7. Outlier Detection:**

• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

• Create scatter plots highlighting outliers in a different color.

• Discuss the significance of these outliers.

---



```python
# Function to detect outliers
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[column] < Q1 - 1.5*IQR) | (df[column] > Q3 + 1.5*IQR)]
    return outliers

# HDI outliers
hdi_outliers_sa = detect_outliers(hdi_south_asia, 'hdi')
hdi_outliers_me = detect_outliers(hdi_middle_east, 'hdi')

# GNI per Capita outliers
gni_outliers_sa = detect_outliers(hdi_south_asia, 'gross_inc_percap')
gni_outliers_me = detect_outliers(hdi_middle_east, 'gross_inc_percap')
```

```python
# Function to plot scatter with outliers, with custom figure label
def scatter_outliers(df, x_col, y_col, region_name, fig_label):
    plt.figure(figsize=(8,5))
    sns.scatterplot(data=df, x=x_col, y=y_col, hue='country', s=70, legend=False)

    # Highlight outliers
    outliers = detect_outliers(df, y_col)
    sns.scatterplot(data=outliers, x=x_col, y=y_col, color='red', s=100, label='Outlier')

    plt.title(f'{fig_label}: {y_col.replace("_"," ").title()} vs {x_col.replace("_"," ").title()} ({region_name})', fontsize=14)
    plt.xlabel(x_col.replace("_"," ").title(), fontsize=12)
    plt.ylabel(y_col.replace("_"," ").title(), fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'Figure_{fig_label}.png', dpi=300)
    plt.show()


# HDI vs GNI per Capita for South Asia → Figure 11a
scatter_outliers(hdi_south_asia, 'gross_inc_percap', 'hdi', 'South Asia', 'Figure 12a')

# HDI vs GNI per Capita for Middle East → Figure 11b
scatter_outliers(hdi_middle_east, 'gross_inc_percap', 'hdi', 'Middle East', 'Figure 12b')
```



*Figure 12a shows HDI vs GNI per Capita for South Asia,* with outliers highlighted in red, indicating countries whose HDI or income levels deviate significantly from the regional trend.

*Figure 12b shows HDI vs GNI per Capita for the Middle East*, highlighting outliers, which reveal countries that perform unusually high or low in HDI relative to their GNI per Capita.

---



**Interpretation:**

The outliers highlight countries that deviate from the typical relationship between HDI and GNI per Capita in their region.

For example, a country with a high GNI per Capita but relatively lower HDI may indicate that wealth is not translating effectively into better health, education, or overall human development.

Conversely, a country with lower GNI but higher HDI suggests effective utilization of limited resources, possibly due to strong social policies or equitable development strategies.

Identifying these outliers helps policymakers and analysts understand exceptional cases and investigate factors contributing to over- or under-performance in human development relative to income.
