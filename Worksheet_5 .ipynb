```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Concept and technologies of AI/Houseprice (1).csv')
print(df.head())
```

```python
age = df['HouseAge'].to_numpy()
floor = df['HouseFloor'].to_numpy()
area = df['HouseArea'].to_numpy()
price = df['HousePrice'].to_numpy()
```

```python
x0 = np.ones(len(age))
X2 = np.array([x0, age, floor, area]).T
W = np.array([0, 0, 0, 0])
Y2 = np.array(price)
```

```python
def cost_function(X, Y, W):
    """
    This function finds the Mean Square Error.
    """
    m = len(Y)
    J = np.sum((X.dot(W) - Y) ** 2) / (2 * m)
    return J
```

```python
# Test case
X_test = np.array([[1, 2], [3, 4], [5, 6]])
Y_test = np.array([3, 7, 11])
W_test = np.array([1, 1])

cost = cost_function(X_test, Y_test, W_test)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong: Reimplement the cost function")
    print("Cost function output:", cost)
```

```python
initial_cost = cost_function(X2, Y2, W)
print(f"Initial Cost: {initial_cost}")
```

```python
def gradient_descent(X, Y, B, alpha, iterations):
    cost_history = [0] * iterations
    m = len(Y)
    # Using B as the weight matrix to update locally
    weights = B.copy().astype(float)

    for iteration in range(iterations):
        Y_pred = X.dot(weights)
        loss = Y_pred - Y
        dw = (X.T.dot(loss)) / m
        weights = weights - alpha * dw

        cost = cost_function(X, Y, weights)
        cost_history[iteration] = cost

    return weights, cost_history
```

```python
alpha = 0.00001
iterations = 100000
new_weights, cost_history = gradient_descent(X2, Y2, W, alpha, iterations)

print(f"New Weights: {new_weights}")
print(f"Final Cost: {cost_history[-1]}")
```

```python
def rmse(Y, Y_pred):
    """ Calculates the Root Mean Square Error. """
    rmse_val = np.sqrt(np.sum((Y - Y_pred)**2) / len(Y))
    return rmse_val
```

```python
def r2(Y, Y_pred):
    """ Calculates the R Squared Error. """
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    r2_val = 1 - (ss_res / ss_tot)
    return r2_val
```

```python
Y_pred = X2.dot(new_weights)

print(f"RMSE: {rmse(Y2, Y_pred)}")
print(f"R2 Score: {r2(Y2, Y_pred)}")
```
